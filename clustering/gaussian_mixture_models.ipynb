```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.mixture import GaussianMixture
```

```python
# Generate example data (blobs dataset)
X, y = make_blobs(n_samples=300, centers=3, random_state=42)
```

```python
# Create the Gaussian Mixture Model
n_components = 3
gmm = GaussianMixture(n_components=n_components)

# Fit the model to the data and obtain cluster assignments
clusters = gmm.fit_predict(X)
```

```python
# Extract the cluster means and covariances
cluster_means = gmm.means_
cluster_covariances = gmm.covariances_

# Print the means and covariances of the Gaussian components
for i in range(n_components):
    print(f"Cluster {i+1} Mean: {cluster_means[i]}")
    print(f"Cluster {i+1} Covariance Matrix:")
    print(cluster_covariances[i])
    print()
```

```python
# Plot the clustering result
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', s=50)
plt.scatter(cluster_means[:, 0], cluster_means[:, 1], c='red', marker='X', s=200, label='Cluster Means')
plt.title("Gaussian Mixture Model Clustering Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
