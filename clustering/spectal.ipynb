```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import SpectralClustering
from sklearn.preprocessing import StandardScaler
```

```python
# Generate example data (moons dataset)
X, y = make_moons(n_samples=1000, noise=0.1, random_state=42)
```

```python
# Scale the data for better clustering performance
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Create the Spectral Clustering model
n_clusters = 2  # Number of clusters to find
spectral_clustering = SpectralClustering(n_clusters=n_clusters, affinity='nearest_neighbors', random_state=42)

# Fit the model to the data and obtain cluster assignments
clusters = spectral_clustering.fit_predict(X_scaled)
```

```python
# Plot the clustering result
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='viridis', s=50)
plt.title("Spectral Clustering Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
