```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import OPTICS
```

```python
# Generate example data (moons dataset)
X, y = make_moons(n_samples=1000, noise=0.1, random_state=42)
```

```python
# Create the OPTICS model
optics = OPTICS(min_samples=5, xi=0.05, min_cluster_size=0.05)

# Fit the model to the data
optics.fit(X)
```

```python
# Extract the cluster labels and core distances
labels = optics.labels_
core_distances = optics.core_distances_
```

```python
# Plot the clustering result
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, edgecolors='k')
plt.title("OPTICS Clustering Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
# Plot the reachability plot
plt.figure(figsize=(8, 6))
plt.plot(np.arange(len(core_distances)), np.sort(core_distances), marker='o')
plt.xlabel("Data Points (sorted by reachability distance)")
plt.ylabel("Reachability Distance")
plt.title("Reachability Plot")
plt.show()
```
