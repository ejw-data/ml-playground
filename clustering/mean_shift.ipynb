```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import MeanShift, estimate_bandwidth
```

```python
# Generate example data (blobs dataset)
X, y = make_blobs(n_samples=300, centers=3, random_state=42)
```

```python
# Estimate the bandwidth (bandwidth controls the size of the kernel used in the algorithm)
bandwidth = estimate_bandwidth(X, quantile=0.3)

# Create the Mean Shift model
meanshift = MeanShift(bandwidth=bandwidth)

# Fit the model to the data and obtain cluster assignments
clusters = meanshift.fit_predict(X)
```

```python
# Extract the unique cluster centers
cluster_centers = meanshift.cluster_centers_

# Print the number of clusters and the estimated cluster centers
n_clusters = len(cluster_centers)
print(f"Number of clusters found: {n_clusters}")
print("Cluster centers:")
print(cluster_centers)
```

```python
# Plot the clustering result
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', s=50)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', marker='X', s=200, label='Cluster Centers')
plt.title("Mean Shift Clustering Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
