```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Generate example data (blobs dataset)
X, y = make_blobs(n_samples=100, centers=3, random_state=42)
```

```python
# Create the AgglomerativeClustering model
n_clusters = 3
agg_clustering = AgglomerativeClustering(n_clusters=n_clusters)

# Fit the model to the data and obtain cluster assignments
clusters = agg_clustering.fit_predict(X)
```

```python
# Plot the clustering result
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', s=50)
plt.title("Agglomerative Hierarchical Clustering Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
```

```python
# Plot the dendrogram (optional)
plt.figure(figsize=(12, 6))
linkage_matrix = linkage(X, method='ward')  # You can change the linkage method
dendrogram(linkage_matrix)
plt.title("Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```
