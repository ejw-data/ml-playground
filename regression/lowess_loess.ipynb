```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
# Generate some synthetic data for the example
np.random.seed(42)
X = np.sort(5 * np.random.rand(100, 1), axis=0)
y = np.sin(X).ravel() + np.random.normal(0, 0.1, X.shape[0])
```

```python
# Lowess smoothing
lowess = sm.nonparametric.lowess(y, X.ravel(), frac=0.3)

# Loess smoothing
loess = sm.nonparametric.lowess(y, X.ravel(), frac=0.7)
```

```python
# Plot the original data and the smoothed curves
plt.figure(figsize=(8, 6))
plt.scatter(X, y, s=20, label="Data")
plt.plot(lowess[:, 0], lowess[:, 1], color='r', label="Lowess Smoothing (frac=0.3)")
plt.plot(loess[:, 0], loess[:, 1], color='g', label="Loess Smoothing (frac=0.7)")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("Lowess and Loess Smoothing")
plt.show()
```
