```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Generate synthetic data with an exponential relationship
np.random.seed(42)
y = np.sort(5 * np.random.rand(100, 1), axis=0)
X = 2 * np.log(0.5 * y) + np.random.normal(0, 0.1, y.shape[0])+10

plt.scatter(X[:,0],y[:,0], s=20, label="Original Data")
```

```python
# Apply logarithmic transformation to X
X_log = np.log(X[:,10])

plt.scatter(X_log,y[:,0], s=20, label="Transformed Y (log(Y))")
```

```python
# Create a Linear Regression model
model = LinearRegression()

# Train the model on the transformed data
model.fit(X_log, y[:,0])

# Make predictions on the transformed data
y_pred = model.predict(X_log)
```

```python
# Plot the original data and the linear regression line on the transformed data
plt.figure(figsize=(8, 6))
plt.scatter(X, y, s=20, label="Original Data")
plt.plot(X, y_pred, color='r', label="Linear Regression on Transformed Data")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("Linear Regression with Logarithmic Transformation")
plt.show()
```
