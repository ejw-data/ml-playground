```python
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df= pd.read_csv("../data/netflix/titles.csv")
df.head()
```

```python
df.fillna('', inplace=True)
```

```python
df['genre_string']=df['genres'].map(lambda x: x.replace("[","").replace("]","").replace("'","").replace(",",""))
df.head()
```

```python
df['text'] = df['title']+" "+df['description']+" "+df['genre_string']
```

```python
df_clean = df[['title', 'text']]
```

```python
df_clean.head()
```

```python
df_clean.reset_index(inplace=True)
```

```python
df_clean.head()
```

```python
cv = CountVectorizer() 
count_matrix = cv.fit_transform(df_clean["text"])
```

```python
cosine_sim = cosine_similarity(count_matrix)
```

```python

def find_title_from_index(index):
    return df_clean[df_clean.index == index]["title"].values[0]

def find_index_from_title(title):
    return df_clean[df_clean['title'] == title]["index"].values[0]
```

```python
show = "Taxi Driver"
show_index = find_index_from_title(show)
```

```python
similar_shows = list(enumerate(cosine_sim[show_index]))
```

```python
sorted_similar_shows = sorted(similar_shows,key=lambda x:x[1],reverse=True)[1:]
```

```python
i=0
for element in sorted_similar_shows:
    print(find_title_from_index(element[0]))
    i=i+1
    if i>5:
        break
```
